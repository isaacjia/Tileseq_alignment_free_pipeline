```python
def seq_to_mutlib_df(tile):
    
    tile_seq = tile_sequence[tile]
    tile_region = (int(tile_pos_dict[tile].split('-')[0]),int(tile_pos_dict[tile].split('-')[1]))
    
    tsnt_dict = {}
    tsnt_dict[tile_seq] = 'tile_nt_WT'
    tsnt_dict['other'] = 'other_nt'
    
    for codon_pos in range(tile_region[0],tile_region[1]+1,1):
        re_pos = codon_pos-tile_region[0]
        left_seq = tile_seq[:3*re_pos]
        right_seq = tile_seq[3*(re_pos+1):]
        codonWT = MSH2_codonWT[codon_pos]
        AA_WT = codon_AA[codonWT]
        for tri_codon in codon64:
            if tri_codon != codonWT:
                tsnt_dict[left_seq + tri_codon + right_seq] = codonWT+'_'+str(codon_pos)+'_'+tri_codon
    
    df = pd.DataFrame.from_dict(tsnt_dict, orient='index')

    df = df.reset_index()
    df.columns = ['seq','var_nt']
    df = df.set_index('seq')
    for seq in df.index:
        if seq not in ['other',tile_seq]:
            var_nt = df.at[seq,'var_nt']
            var_AA = codon_AA[var_nt.split('_')[0]]+'_'+var_nt.split('_')[1]+'_'+ codon_AA[var_nt.split('_')[2]]
            df.at[seq,'var_AA'] = var_AA
        elif seq == 'other':
            df.at[seq,'var_AA'] = 'other_AA'
        elif seq == 'tile_nt_WT':
            df.at[seq,'var_AA'] = 'tile_AA_WT'
    return df
```
